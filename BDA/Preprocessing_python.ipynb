### Dataset Used: **India_GDP**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/India_GDP_1960-2022.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.columns=data.iloc[0,:]
```

```python
data.sort_values(by=['Year'])
data.drop(0,inplace=True)
```

```python
data=data.sort_values(by=['Year'])
```

```python
data=data.astype({'GDP in (Billion) $': 'float', 'Per Capita in rupees': 'float','Growth %': 'float'})
```

```python
data.set_index('Year',inplace=True)
```

```python
data[data['GDP in (Billion) $']<100]
```

```python
data[data['Growth %']<=3]
```

**Simple Visualization**

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(data['GDP in (Billion) $'],color='g',kde=True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(data['Per Capita in rupees'],color='y',kde=True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(data['Growth %'],color='r',kde=True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=data,x=data.index,y='Per Capita in rupees',ax=ax,color='peru')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=data,x=data.index,y='GDP in (Billion) $',ax=ax,color='k')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=data,x=data.index,y='Growth %',ax=ax,color='navy')
plt.show()
```
